# Data Visualization

## ETT

### ETTh1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('ETTh1.csv')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the 'date' column as the index
data.set_index('date', inplace=True)

# Plot the data
plot_list = [['HUFL', 'HULL'], ['MUFL', 'MULL'], ['LUFL', 'LULL'], ['OT']]
fig = plt.figure(figsize=(10, 10))
for i in range(4):
    fig.add_subplot(2, 2, (i+1))
    for j in range(len(plot_list[i])):
        plt.plot(data[plot_list[i][j]], label=plot_list[i][j])
    plt.xlabel('Date')
    plt.ylabel('Temperature')
    # # Add title
    plt.title('ETTh1 Temperature Readings')
    plt.legend()
    

# Show the plot
plt.show()
```

## WTH

```python
# Load the data
data = pd.read_csv('WTH.csv')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the 'date' column as the index
data.set_index('date', inplace=True)

# Plot the data
plot_list = ['RelativeHumidity','WindSpeed','WindDirection','StationPressure','Altimeter','WetBulbCelsius']
fig = plt.figure(figsize=(12, 7))
plt.title('Weather Dataset Temperature Readings')
for i in range(6):
    fig.add_subplot(2, 3, (i+1))
    plt.plot(data[plot_list[i]], label=plot_list[i])
    plt.xlabel('Date')
    plt.ylabel(plot_list[i])

# Show the plot
plt.show()
```

## ECT

```python
# Load the data
data = pd.read_csv('ECL.csv')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the 'date' column as the index
data.set_index('date', inplace=True)

# Select the first 10 time series
selected_data = data.iloc[:, :10]

# Create the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(selected_data.transpose(), cmap='viridis')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Time Series')
plt.title('ECL Dataset Heatmap')

# Show the plot
plt.show()
```
